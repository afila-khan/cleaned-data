Email Address Extraction
---

```python
import re
def email_extractor(n):
    word=n.lower()
    email = re.findall(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b',word)
    if len(email)==0:
        # return "Not Available"
        return " "
    else:
        return email[0]
merged_df['affiliation']=merged_df['affiliation'].astype(str)
merged_df['Author_Email']=merged_df['affiliation'].apply(email_extractor)
merged_df     
```

Removing Email Addresses from Affiliations
---

```python
def email_remover(n):
    word=n
    if "@" in word:
        word=n.lower()
        if "electronic address" in word:
            word=re.sub('electronic address:', '', word)
            word=re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', '',word)
        else:
            word=re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', '',word)
    return word
merged_df['affiliation']=merged_df['affiliation'].astype(str)
merged_df['affiliation']=merged_df['affiliation'].apply(email_remover)
merged_df[merged_df['affiliation'].astype(str).str.contains("@")]
        
```

```python
country_list=[]
for i in list(pcy.countries):
    country_list.append((i.name).lower())
    try:
        country_list.append((i.official_name).lower())
    except AttributeError:
        country_list.append((i.name).lower())
country_list.append("uk")
country_list.append("us")
country_list.append("usa")
country_list.append("republic of korea")
list1 = usual_countries_list
list2 = country_list

# Add elements from list1 that are not in list2 to list2
list2.extend([item for item in list1 if item not in list2])
list2 = list(set(list2)) # Removing duplicates
all_country_list=list2
all_country_list
```

```python
from fuzzywuzzy import fuzz
def country_remover(n):
    word_split=n.split(",")
    lastcountry=word_split[-1].replace(" ", "")
    likely_country=[]
    similarity_scores=[]
    for i in all_country_list:
        similarity = fuzz.ratio(lastcountry,i)
        if similarity>70:
            likely_country.append(i)
            similarity_scores.append(similarity)
    if len(similarity_scores)!=0:    
        return likely_country[similarity_scores.index(max(similarity_scores))]
    else:
        # return "Not Available"
        return " "
            
merged_df['Country']=merged_df['affiliation'].apply(country_remover)
merged_df
```

```python
def department_extractor(n):
    word=n.split(",")
    department=[]
    for i in word:
        if 'department' in i or 'school' in i:
            department.append(i)
    if len(department)!=0:
        return  str(department)[1:-1]
    else:
        # return "Not Available"
        return " "
merged_df['Department']=merged_df['affiliation'].apply(department_extractor)
merged_df       
```

Province Extractor
---

```python
from countryinfo import CountryInfo
countries_list = [CountryInfo(country).info()['name'] for country in CountryInfo().all()]
usual_countries_list=[]
for i in countries_list:
    usual_countries_list.append(i.lower())
usual_countries_list

province_list=[]
for country_name in usual_countries_list:
    country = CountryInfo(country_name)
    provinces = country.provinces()
    for province in provinces:
        province_list.append(province.lower())

def province_extractor(word):
    word=word.split(",")
    final_province=[]
    for i in word:
        if i in province_list:
            final_province.append(i)
    if len(final_province)!=0:
        return str(list(set(final_province)))[1:-1]
    else:
        # return "Not Available"
        return " "
    
merged_df['Province_or_State']=merged_df['affiliation'].apply(province_extractor)
merged_df    
```
